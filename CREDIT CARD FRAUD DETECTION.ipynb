```python
legit_sample = legit.sample(n=492)
```

```python
data= pd.concat([legit_sample, fraud], axis=0)
```

```python
x= data.drop(columns=['Class'],axis=1)
```

```python
y= data['Class']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, stratify=y, test_size=0.2, random_state=49)
```

```python
model= LogisticRegression( max_iter= 400)
```

```python
model.fit(x_train, y_train)
```

```python
train_pred= model.predict(x_train)
```

```python
acc= accuracy_score(train_pred, y_train)
```

```python
print('Accuracy of training data :', acc)
```


# model evaluation

```python
test_pred= model.predict(x_test)
acc2= accuracy_score(test_pred, y_test)
print('Accuracy of training data :', acc2)
```
